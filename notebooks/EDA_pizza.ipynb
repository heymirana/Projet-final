```python
# comme dit Antonin dataframe avec jour mois et revenue 
df_grouped_revenue = df_pizza.groupby(['month', 'day_of_month']).agg({'revenue': 'sum'}).reset_index()
```

```python
# graphique CA au globale pour chaque mois 
plt.figure(figsize=(10, 6)) 
sns.lineplot(data=df_grouped_revenue, x='day_of_month', y='revenue', hue='month', palette='tab10')

plt.xlabel('Jour du mois')
plt.ylabel('Chiffre d\'affaires')
plt.title('Chiffre d\'affaires par jour pour chaque mois')

plt.legend(title='Mois')
plt.show()
""" dans ce graphique j'ai voulu voi, au global qu'est ce que ça donne ? """
```

```python
#  4 sous-graphiques : 2 lignes et 2 colonnes
fig, axes = plt.subplots(2, 2, figsize=(14, 10))  

# avoir les données en trimestre
trimestres = {
    'Janvier-Mars': df_grouped_revenue[df_grouped_revenue['month'].isin([1, 2, 3])],
    'Avril-Juin': df_grouped_revenue[df_grouped_revenue['month'].isin([4, 5, 6])],
    'Juillet-Septembre': df_grouped_revenue[df_grouped_revenue['month'].isin([7, 8, 9])],
    'Octobre-Décembre': df_grouped_revenue[df_grouped_revenue['month'].isin([10, 11, 12])]
}

for ax, (trimestre, data) in zip(axes.flatten(), trimestres.items()):
    sns.lineplot(data=data, x='day_of_month', y='revenue', hue='month', palette='tab10', ax=ax)
    ax.set_title(f'Chiffre d\'affaires {trimestre}')
    ax.set_xlabel('Jour du mois')
    ax.set_ylabel('Chiffre d\'affaires')
    ax.legend(title='Mois')

# Ajuster les espacements pour éviter que les sous-graphiques se chevauchent
plt.tight_layout()
plt.show()
```

```python
df_pizza['order_date'] = pd.to_datetime(df_pizza['order_date'])
```

```python
"""from math import ceil

#fonction pour avoir le numéro de semaine 
def number_of_week(dt):
    first_day = dt.replace(day=1) # le premier jour du mois
    day_of_the_month = dt.day # jour
    adjusted_day = day_of_the_month + first_day.weekday()

    return int(ceil(adjusted_day/7.0)) # l'équivalent à  (adjusted_dom - 1) // 7 + 1"""
def number_of_week(date):
    # Premier jour du mois
    first_day = date.replace(day=1)
    
    # Calculer le numéro de semaine en fonction du jour de la semaine du premier jour
    # On considère que la semaine commence le lundi (0 = lundi, 6 = dimanche)
    week_start = first_day.weekday()  # 0 = lundi, 6 = dimanche
    
    # Calculer le numéro de semaine
    week_number = (date.day + week_start - 1) // 7 + 1
    
    return week_number
```

```python
# je crée la colonne week_number
df_pizza['week_number'] = df_pizza['order_date'].apply(number_of_week)
```

```python
# je crée un nouveau dataframe où je mets le nb de jour , le numéro de semaine et l'aggreg
# comme dit Antonin 
df_grouped_revenue_week = df_pizza.groupby(['day_of_week', 'week_number']).agg({'revenue': 'sum'}).reset_index()
```
